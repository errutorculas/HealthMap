```python
import numpy as np
import pandas as pd

import config
import requests         #for Geocoding API

%matplotlib inline
```

```python
df = pd.read_csv ('datasets/pharmacy-healthmap.csv')

df
```

Geocoding Latitudes and Longitudes

```python
def fetchLatLng(location):
    address = location

    params = {
        'key' : config.API_KEY,
        'address' : address
    }

    response = requests.get(config.base_url, params=params).json()

    # print(response['results'])

    if response['status'] == 'OK':
        geometry = response['results'][0]['geometry']
        lat = geometry['location']['lat']
        lng = geometry['location']['lng']

    return(lat, lng)
```

```python
for i, row in df.iterrows():
    coordinates = fetchLatLng(row['address'])

    #print(coordinates)

    df.at[i,'lat'] = coordinates[0]
    df.at[i,'long'] = coordinates[1]

```
