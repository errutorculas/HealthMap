```python
import numpy as np
import pandas as pd

%matplotlib inline
```

```python
dfASC = pd.read_csv ('datasets/csv/DOH-AmbulatorySurgicalClinic.csv')
dfDia = pd.read_csv ('datasets/csv/DOH-DialysisClinic.csv')

dfClinicLab = pd.read_csv ('datasets/csv/DOH-ClinicalLaboratory.csv')
dfClinicLab
```

```python
filtered_ASC = dfASC[dfASC["REGION"].str.contains("REGION 6", na=False)]
filtered_Dia = dfDia[dfDia["REGION"].str.contains("REGION 6", na=False)]

filtered_Clinic = dfClinicLab[dfClinicLab["Region"].str.contains("REGION 6", na=False)]

filtered_Clinic
```

```python
toCSV_ASC = filtered_ASC[['NAME OF FACILITY', 'COMPLETE ADDRESS']]
toCSV_Dia = filtered_Dia[['NAME OF CLINIC', 'COMPLETE ADDRESS']]

toCSV_Clinic = filtered_Clinic[['Name of Facility', 'Province', 'City/Municipality', 'Address']]
```

```python
toCSV_ASC.to_csv('ASC-Clinic.csv', index=False)
toCSV_Dia.to_csv('Dia-Clinic.csv', index=False)

toCSV_Clinic.to_csv('ClinicLab.csv', index=False)
```
